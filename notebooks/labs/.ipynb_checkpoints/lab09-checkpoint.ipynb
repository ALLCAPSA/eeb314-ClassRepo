Here let's look at the so-called Rosenzweig-MacArthur model of predator and prey.

This model assumes that the density of prey, $N$, grows logistically in the absence of predators, with per capita growth rate $r$ and carrying capacity $k$. Predators, with density $P$, consume prey at a per prey per predator rate of $a/(1+b N)$. The reason we assume this declines with prey density is that consuming prey takes time. Finally, we assume each consumed prey is converted into a fraction $c$ new predators, and that predators die at per capita rate $d$. Putting all this together in a continuous-time model we have

$\frac{\mathrm{d}N}{\mathrm{d}t} = r N (1 - N / k) - \frac{a}{1 + b N} N P$

$\frac{\mathrm{d}P}{\mathrm{d}t} = c\frac{a}{1 + b N} N P - d P$

Here is a little SymPy to get you going

````python
from sympy import * #load symbolic library

var('r,N,k,a,b,P,c,d') #define variables and parameters

dNdt = r*N*(1-N/k) - a/(1+b*N)*N*P #differntial equation for the prey
dPdt = c*a/(1+b*N)*N*P - d*P #differential equation for the predator
````

## Phase plane analysis

Here I'll just show you what this looks like on a phase plane, as there is a nice connection between the stability analysis above and the nullclines in this model. First we define some functions

````python
import matplotlib.pyplot as plt #import plotting library
import numpy as np #import numeric library
````

````python
def plotnullclines(dNdt, dPdt, pvals, Nmax=10, Pmax=5):
    '''plot nullclines'''

    Ns = np.linspace(0,Nmax,100)
    Ps = np.linspace(0,Pmax,100)

    fig, ax = plt.subplots()

    NnullP = solve(dNdt,P)
    ax.plot(Ns, [NnullP[0].subs(pvals).subs(N,i) for i in Ns]) #N null cline

    PnullN = solve(dPdt,N)
    ax.plot([PnullN[0].subs(pvals) for _ in Ps], Ps) #P null cline

    ax.set_ylim(0,Pmax)

    return fig, ax
````

````python
def NP(N0, P0, dNdt, dPdt, pvals, dt=0.1, max=np.inf):
    '''generator for dynamics'''
    
    t, Nt, Pt = 0, N0, P0 #initial conditions
    while t < max:
        yield Nt, Pt #current state
        t += dt #update t
        Nt += dt*dNdt.subs(pvals).subs(N,Nt).subs(P,Pt) #update N
        Pt += dt*dPdt.subs(pvals).subs(N,Nt).subs(P,Pt) #update P
````

Now, choose some parameter values, with a small carrying capacity $k$. We see that the predator null cline (orange) meets the prey null cline (blue) on it's descending part. In this case the dynamics from a given starting point (black) go to the coexistence equilibrium, ie, the equilibrium is stable.

````python
pvals = {'k':10, 'r':0.1, 'a':0.05, 'b':0.1, 'c':0.1, 'd':0.02}
fig, ax = plotnullclines(dNdt=dNdt, dPdt=dPdt, pvals=pvals)

NP1 = NP(5,1,dNdt,dPdt,pvals,dt=1,max=1000)
NP1t = np.array([i for i in NP1])
ax.plot(NP1t[:,0],NP1t[:,1], c='k')

plt.show()
````

Now increase $k$ and we see that now the predator nullcline intersects with the prey nullcline on its ascending part. And now the coexistence equilibrium is unstable.

````python
pvals = {'k':100, 'r':0.1, 'a':0.05, 'b':0.1, 'c':0.1, 'd':0.02}
fig, ax = plotnullclines(dNdt=dNdt, dPdt=dPdt, pvals=pvals, Nmax=100, Pmax=10)

NP1 = NP(5,3,dNdt,dPdt,pvals,dt=1,max=1000)
NP1t = np.array([i for i in NP1])
ax.plot(NP1t[:,0],NP1t[:,1], c='k')

plt.show()
````

It turns out that the in this system the slope of the prey nullcline has the opposite sign as the trace of the Jacobian at the coexistence equilibrium. Therefore whenever the predator nullcline is to the right of the maximum of the prey null cline then coexistence is stable, and vice versa. And since increasing $k$ moves the maximum of the prey nullcline to the right, it destabilizes the system.
